```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### **Info General**

```python
datos = pd.read_csv("data/Pokemon.csv",  index_col="Number")
datos # Es la A
```

```python
datos.describe() # Es la A
```

#### **Para ver los duplicados**

```python
ssss=datos.duplicated().sum()
print(ssss) #Mostrando cantidad de duplicados
```

```python

datos_no_duplicados=datos.drop_duplicates()#Elimina los duplicados
datos_no_duplicados
```

#### **Elimina los Datos Nulos**

```python
datos_nulos = datos_no_duplicados[datos_no_duplicados.isnull().any(axis=1)] #Esta es la C
datos_nulos #Muestra solo los datos nulos   
```

```python
datos_nulos = datos_no_duplicados.dropna() #Esta es la C
datos_nulos #Aqui se muestran los datos sin los nulos
```

#### **Filtros**

```python
def filtro_fuego(Tipo_Fuego):
    return 'Fire' in Tipo_Fuego

Fuego_1 = datos_nulos[datos_nulos['Type 1'].apply(filtro_fuego)]
Fuego_Total_1=Fuego_1['Type 1']
Fuego_Total_1.head(20)
```

```python
def filtro_fuego(Tipo_Fuego):
    return 'Fire' in Tipo_Fuego

Fuego_2 = datos_nulos[datos_nulos['Type 2'].apply(filtro_fuego)]
Fuego_Total_2=Fuego_2['Type 2']
Fuego_Total_2.head(20)
```

```python
def filtro_fuego(Tipo_Fuego):
    return 'Def' in Tipo_Fuego

Fuego_1 = datos_nulos[datos_nulos['Def'].apply(filtro_fuego)]
Fuego_Total_1=Fuego_1['Def']
Fuego_Total_1.head(20)
```
