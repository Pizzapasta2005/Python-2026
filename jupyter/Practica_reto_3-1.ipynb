```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def procesar_caudal():
    horas = np.arange(1,9) # array = [1,2,3,4,5,6,7,8]
    caudales = [12.5, 14.2, 18.1, 22.8, 25.4, 21.0, 17.5, 15.3]
    #Creando diccionario 
    data = {
        'Hora': horas,
        'Caudal': caudales,
    }

    #transformando el diccionario a dataframe
    df = pd.DataFrame(data)
    return df

```

```python
def visualizar_curva(dataframe):
    fig, ax = plt.subplots() # -> ax = axes / fig  = figure

    
    ax.plot(dataframe['Hora'], dataframe['Caudal'], marker='o', color='green', linestyle= '--')
    ax.set_title('Monitoreo de Caudal Rio Gamboa')
    ax.set_xlabel('Tiempo (Horas)')
    ax.set_label('Caudal (m^3/s)')



    ax.grid(True)
```

```python
if __name__ == "__main__":
    dat = procesar_caudal()
    visualizar_curva(dat)
    plt.show()
```
